### Плоский JSON-объект

#### Константы

```python
axum_df = pd.read_csv(f"{DMES_PATH}/axum_flat.csv")
actix_df = pd.read_csv(f"{DMES_PATH}/actix_flat.csv")

axum_users = axum_df["users_cnt"]
actix_users = actix_df["users_cnt"]
```

#### Время ответа

```python
fig = plt.time_cmp_plot(
    (axum_users, axum_df["response_time"]),
    (actix_users, actix_df["response_time"])
)
fig.show()
```

#### CPU

```python
fig = plt.plot_resources_cmp_graph(
    axum_data=(axum_users, (axum_df['min_cpu_percent'], axum_df['med_cpu_percent'], axum_df['max_cpu_percent'])),
    actix_data=(actix_users, (actix_df['min_cpu_percent'], actix_df['med_cpu_percent'], actix_df['max_cpu_percent'])),
    y_label="CPU %"
)
fig.show()
```

#### RAM

```python
fig = plt.plot_resources_cmp_graph(
    axum_data=(axum_users, (axum_df['min_ram_percent'], axum_df['med_ram_percent'], axum_df['max_ram_percent'])),
    actix_data=(actix_users, (actix_df['min_ram_percent'], actix_df['med_ram_percent'], actix_df['max_ram_percent'])),
    y_label="RAM %"
)
fig.show()
```
